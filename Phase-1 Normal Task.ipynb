**Exploratory Data Analysis - Terrorism**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
M=pd.read_csv("terrorism.csv")
M
```

```python
M = M.drop(["Unnamed: 0", "eventid"], axis=1)
M=M.dropna()
M=M.drop_duplicates()
M
```

```python
print(M.describe())
```

```python
print(M.nunique())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=M, x="year")
plt.title("Number of Attacks per Year")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(data=M, x="country", order=M["country"].value_counts().index[:10])
plt.title("Top 10 Countries with the Most Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=M, x="motive", order=M["motive"].value_counts().index[:10])
plt.title("Top 10 Attack Motives")
plt.xticks(rotation=90)
plt.show()
```

```python
# Property Damage Analysis
plt.figure(figsize=(7, 4))
sns.countplot(data=M, x="property_extent_damaged")
plt.title("Extent of Property Damage")
plt.xticks(rotation=90)
plt.show()
```

```python
# Target Analysis
plt.figure(figsize=(7, 4))
sns.countplot(data=M, x="nationality_targeted", order=M["nationality_targeted"].value_counts().index[:10])
plt.title("Top 10 Targeted Nationalities/Groups")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=M, x="weapon_type", order=M["weapon_type"].value_counts().index[:10])
plt.title("Top 10 Weapon Types Used in Attacks")
plt.xticks(rotation=45)
plt.show()
```
